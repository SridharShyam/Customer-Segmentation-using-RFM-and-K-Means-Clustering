```python
from sklearn.preprocessing import StandardScaler

to_scale=rfm[['Recency', 'Frequency', 'Monetary']]
scaler=StandardScaler()
rfm_scaled=scaler.fit_transform(to_scale)

rfm_scaled=pd.DataFrame(rfm_scaled,columns=['Recency', 'Frequency', 'Monetary'])

rfm_scaled
```

```python
from sklearn.cluster import KMeans

inertia = []
k_values = range(1, 11)

for k in k_values:
    k_means = KMeans(n_clusters=k, random_state=42)
    k_means.fit(rfm_scaled)
    inertia.append(k_means.inertia_)

plt.figure(figsize=(6, 5))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal K')
plt.show()
```

```python
k_optimal = 4
k_means = KMeans(n_clusters=k_optimal, random_state=42)

rfm = rfm.copy()

rfm['Clusters'] = k_means.fit_predict(rfm_scaled)

cluster_summary = rfm.groupby('Clusters')[['Recency', 'Frequency', 'Monetary']].mean().round(4)

cluster_summary
```

```python
from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(rfm_scaled, rfm['Clusters'])
print(f"Davies-Bouldin Index: {db_score:.3f}")
```

```python
import pandas as pd

# Make sure rfm_scaled is a DataFrame with CustomerID included
rfm = rfm_scaled.copy()

# Use rank to simulate RFM scoring
rfm['R_Score'] = rfm['Recency'].rank(pct=True)
rfm['F_Score'] = rfm['Frequency'].rank(pct=True)
rfm['M_Score'] = rfm['Monetary'].rank(pct=True)

# Convert to 1â€“5 score buckets
rfm['R_Score'] = pd.cut(rfm['R_Score'], bins=5, labels=[5, 4, 3, 2, 1]).astype(int)
rfm['F_Score'] = pd.cut(rfm['F_Score'], bins=5, labels=[1, 2, 3, 4, 5]).astype(int)
rfm['M_Score'] = pd.cut(rfm['M_Score'], bins=5, labels=[1, 2, 3, 4, 5]).astype(int)

# Combine scores
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)

# Segment customers
def rfm_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']
    if r >= 4 and f >= 4 and m >= 4:
        return 'Loyal Customers'
    elif r >= 4 and f >= 4 and m < 4:
        return 'Potential Loyalists'
    elif r >= 4 and f < 4 and m >= 4:
        return 'Big Ticket Shoppers'
    elif r >= 4 and f < 4 and m < 4:
        return 'New or Inactive Customers'
    elif r < 4 and f >= 4 and m >= 4:
        return 'At-Risk High-Value Customers'
    elif r < 4 and f >= 4 and m < 4:
        return 'Frequent Low Spenders'
    elif r < 4 and f < 4 and m >= 4:
        return 'At-Risk Big Spenders'
    else:
        return 'Lost or Dormant Customers'

rfm['RFM_Category'] = rfm.apply(rfm_segment, axis=1)

# Display result with CustomerID and other columns
rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary', 'R_Score', 'F_Score', 'M_Score', 'RFM_Category']]
```

```python
import matplotlib.pyplot as plt

segment_counts = rfm['RFM_Category'].value_counts()

plt.figure(figsize=(10, 6))
bars = plt.barh(segment_counts.index, segment_counts.values, color='#00bfff', edgecolor='black')

plt.title('Customer Segment Distribution', fontsize=16, weight='bold')
plt.xlabel('Number of Customers', fontsize=12)
plt.ylabel('Customer Segment', fontsize=12)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2,
             f'{int(width)}', va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
